```python
class Perfume:
    def __init__(self, nombre, marca, fragancia):
        self.nombre = nombre
        self.marca = marca
        self.fragancia = fragancia
    def leer_datos(self):
        print(f"\n[SISTEMA]: Registrando el perfume '{self.nombre}' exitosamente.")
    def mostrar_datos(self):
        print(f"\n--- FICHA TÉCNICA ---")
        print(f"Nombre:    {self.nombre}")
        print(f"Marca:     {self.marca}")
        print(f"Fragancia: {self.fragancia}")
        print(f"----------------------")
    def aplicar_fragancia(self):
        print(f" >> ¡Bshhh! El spray de {self.marca} se ha activado.")
        print(f" >> Nota mental: Huele intensamente a {self.fragancia}.")
# --- Bloque Interactivo ---
print("=== BIENVENIDO AL REGISTRO DE PERFUMERÍA ===")
# Registro del primer perfume
print("\nIngresa los datos del PERFUME 1:")
nom1 = input("Nombre: ")
mar1 = input("Marca: ")
fra1 = input("Fragancia: ")
perfume1 = Perfume(nom1, mar1, fra1)
# Registro del segundo perfume
print("\nIngresa los datos del PERFUME 2:")
nom2 = input("Nombre: ")
mar2 = input("Marca: ")
fra2 = input("Fragancia: ")
perfume2 = Perfume(nom2, mar2, fra2)
# Ejecución de acciones para el Perfume 1
perfume1.leer_datos()
perfume1.mostrar_datos()
perfume1.aplicar_fragancia()
# Ejecución de acciones para el Perfume 2
perfume2.leer_datos()
perfume2.mostrar_datos()
perfume2.aplicar_fragancia()
print("\n=== PROCESO FINALIZADO ===")
print("Programa creado por Daniel Nieves NL = 41 ")
```
